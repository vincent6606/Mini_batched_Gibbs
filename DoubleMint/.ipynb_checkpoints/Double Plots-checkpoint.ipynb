```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("paper", font_scale=1.5)
flatui = ["#0a4d8c", "#ba0c2f", "#49a343", "#6E10FF", "#DE4C0B", "#99583D", "#4d2177", "#7c3a2d"]
# https://color.adobe.com/zh/create/color-wheel/?basfrom matplotlib import rc
sns.palplot(flatui)
```

```python
def iter_loadtxt(filename, delimiter=',', skiprows=0, dtype=float):
    def iter_func():
        with open(filename, 'r') as infile:
            for _ in range(skiprows):
                next(infile)
            for line in infile:
                line = line.rstrip().split(delimiter)
                for item in line:
                    yield dtype(item)
        iter_loadtxt.rowlength = len(line)

    data = np.fromiter(iter_func(), dtype=dtype)
    data = data.reshape((-1, iter_loadtxt.rowlength))
    return data
```

```python
error = iter_loadtxt('DoubleMin_errors_10_10_100000.csv')
MGPMH_error = iter_loadtxt('MGPMH_errors_10_10_100000.csv')
gibbs_error = iter_loadtxt('Gibbs_errors_10_10_100000.csv')
# gibbs_error2 = iter_loadtxt('MGPMH_errors_10_10_100000_new.csv')
```

```python
error[:,0]
```

```python
error = error[error[:,0].argsort()[::1]]
MGPMH_error = MGPMH_error[MGPMH_error[:,0].argsort()[::1]]
```

```python
error[:,0]
```

```python
index = np.arange(100,100000,100)
```

```python
7*1.618
```

```python
import matplotlib.lines as lines
with sns.axes_style("white"):
    fig = plt.figure(figsize=(8.5, 6))
    ax  = fig.add_subplot(111)
    plt.yscale('log')
    plt.xscale('log')

    for i in range(3):
        line = error[i,index]
        plt.plot(index,line,color=flatui[i],linestyle='--',linewidth=2,label="$\lambda = $ {:d} $\psi$".format(i+1))
    
    for i in [3]:
        line = MGPMH_error[i,index]
        plt.plot(index,line,color=flatui[3],linestyle=':',linewidth=2.5,label="MGPMH $\lambda = $ {:d}".format(int(MGPMH_error[i,0])))
    
#     for i in [1]:
#         line = gibbs_error2[0,index]
#         plt.plot(index,line,color=flatui[-2],linestyle=':',linewidth=2,label="MGPMH $\lambda = $ {:d}".format(int(MGPMH_error[i,0])))
    
    for i in range(1):
        line = gibbs_error[0,index]
        plt.plot(index,line,color=flatui[-1],linewidth=2,label="Gibbs Sampling".format(i))
    
    
    plt.ylim(0.03,1)
    plt.yticks((0.03, 0.1, 1), ('0.01', '0.1', '1'), color='k', size=15)
    plt.xticks(color='k',size=15)
    plt.grid(True, which="both")
    plt.xlabel("iterations", size=20)
    plt.ylabel("error in marginals", size=20)
    plt.title(r'DoubleMIN vs. Gibbs sampling ($\beta$ = 6.67)',size=21)
    plt.legend(loc='lower left',fontsize=20)
    plt.grid(linestyle='dotted')
    plt.tight_layout(pad=1, w_pad=1, h_pad=1)
    
    plt.savefig("Marginals_{}.pdf".format('50'), bbox_inches='tight')
    plt.show()
```
