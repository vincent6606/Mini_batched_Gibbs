```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("paper", font_scale=1.5)
flatui = ["#0a4d8c", "#ba0c2f", "#49a343", "#6E10FF", "#DE4C0B", "#99583D", "#4d2177", "#7c3a2d"]

# https://color.adobe.com/zh/create/color-wheel/?basfrom matplotlib import rc
sns.palplot(flatui)
```

```python
error = np.genfromtxt('MINT_Ising_Dep_20_1000000.csv',delimiter=',')
```

```python
error.shape
```

```python
gibbs = np.genfromtxt('Gibbs_Ising_Dep_20_1000000.csv',delimiter=',')
```

```python
import matplotlib.lines as lines


with sns.axes_style("white"):
    fig = plt.figure(figsize=(8, 5.5))
    ax  = fig.add_subplot(111)

    for i in range(error.shape[0]):
        line = error[i,1:]
        plt.plot(line,color=flatui[i],label="$\lambda = $ {0:.4f}".format(error[i,0]))
    for i in range(1):
        gline = gibbs[0,1:]
        plt.plot(gline,color=flatui[i+4],label="Gibbs sampling".format(i))

    plt.xlabel("iterations", size=15)
    plt.ylabel("Error in Marginals", size=15)
    plt.title('MINT Gibbs vs. Gibbs sampling')
    plt.legend(loc='upper right')
    plt.grid(linestyle='dotted')
    plt.tight_layout(pad=1, w_pad=1, h_pad=1)
    
    plt.savefig("Marginals_{}.pdf".format('50'), bbox_inches='tight')
    plt.show()
```
