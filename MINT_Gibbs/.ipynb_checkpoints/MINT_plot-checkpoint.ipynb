```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("paper", font_scale=1.5)
flatui = ["#0a4d8c", "#ba0c2f", "#49a343", "#6E10FF", "#DE4C0B", "#99583D", "#4d2177", "#7c3a2d"]

# https://color.adobe.com/zh/create/color-wheel/?basfrom matplotlib import rc
sns.palplot(flatui)
```

```python
error = np.genfromtxt('MIN_Ising_Dep_20_100000.csv',delimiter=',')
```

```python
gibbs_error = np.genfromtxt('Gibbs_MIN_Ising_20_100000.csv',delimiter=',')
```

```python
error[:,0]
psi= [1,2,4]
```

```python
error = error[error[:,0].argsort()[::1]]
```

```python
index = np.arange(10,100000,1)
```

```python
import matplotlib.lines as lines


with sns.axes_style("white"):
    fig = plt.figure(figsize=(8.5, 6))
    ax  = fig.add_subplot(111)
    plt.yscale('log')
    plt.xscale('log')

    for i in range(error.shape[0]):
        line = error[i,1:]
        plt.plot(index,line[index],linestyle='--',linewidth=2.5,color=flatui[i+1],label="$\lambda = $ {:d} $\psi$".format(int(np.ceil(psi[i]))))

    for i in range(1):
        gline = gibbs_error[1:]
        plt.plot(index,gline[index],color=flatui[i],linewidth=2.5,label="Gibbs sampling".format(i))
    plt.ylim(0.01,1)
    plt.yticks((0.01, 0.1, 1), ('0.01', '0.1', '1'), color='k', size=15)
    
    plt.xticks(color='k',size=15)
    plt.grid(True, which="both",axis='y')
    plt.xlabel("iterations", size=20)
    plt.ylabel("error in marginals", size=20)
    plt.title(r'MIN-Gibbs vs. Gibbs ($\beta$=1.428) ',size=21)
    plt.legend(loc='lower left',fontsize=20)
    plt.grid(linestyle='dotted')
    plt.tight_layout(pad=1, w_pad=1, h_pad=1)
    
    plt.savefig("Marginals_{}.pdf".format('50'), bbox_inches='tight')
    plt.show()
```
